```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
nba = pd.read_csv('nba.csv', index_col = 0)
```

```python
#Q1
nba.isna().any()
```

```python
nba.head()
```

```python
#Q2
nba['Date'] = pd.to_datetime(nba['Date'])
```

```python
nba['Date'].dtype
```

```python
grouped = nba.groupby([nba['Team'], nba['Date'].dt.year])[['TeamPoints', 'OpponentPoints']]
```

```python
grouped.mean()
```

```python
#Q3
nba_2017 = nba[nba['Date'].dt.year == 2017]
nba_2017
```

```python
#Q4
bar_data = nba_2017.groupby('Team')['TeamPoints'].mean()
ax =bar_data.plot.bar(figsize=(12,8))

ax.set_xlabel('Teams in 2017')
ax.set_ylabel('Average Team Points')
ax.set_title('Bar Chart Of Average Team points in 2017', size=20)
```

```python
#Q5
Q5 = nba_2017.groupby('Team')['OpponentPoints'].mean()
Q5.sort_values().iloc[0:5]
```

```python
#Q6
Q6 = pd.crosstab(nba_2017['Team'], nba_2017['WINorLOSS'])
ax = Q6.plot.barh(stacked= True, figsize=(12,8))
ax.set_xlabel('Win and Loss points')
ax.set_ylabel('Teams')
ax.set_title('Bar Chart Of W & L points in 2017', size=20)
```

```python
#Q7
nba_2017['diff'] = nba_2017['TeamPoints'] - nba_2017['OpponentPoints']
nba_2017.head()
```

```python
diff = nba_2017.groupby('Team')['diff'].mean()
```

```python
diff.sort_values(ascending=False)
```

```python
#Q8
houstonData = nba_2017[nba_2017['Team'] == 'HOU']
sns.scatterplot(x=houstonData['TeamPoints'],y=houstonData['OpponentPoints'], data=houstonData)
plt.xlabel('Team points')
plt.ylabel('Opponent Points')
plt.title('Scatter Plot of Houston Rocket’s games’ TeamPoints and OpponentPoints', size=15)
sns.set(rc={'figure.figsize':(11.7,8.27)})
```

```python
#Q9
count_ct = pd.crosstab(houstonData['WINorLOSS'], houstonData['Home'])
count_ct
```

```python
count_ct.loc['W', 'Away'] / (count_ct.loc['W', 'Away'] + count_ct.loc['L', 'Away'])
```

```python
count_ct.loc['W', 'Home'] / (count_ct.loc['W', 'Home'] + count_ct.loc['L', 'Home'])
```

```python
#Q10
groupedByYear = nba.groupby(nba['Date'].dt.year).mean()
groupedByYear=groupedByYear[['FieldGoals.', 'X3PointShots.', 'FreeThrows.']]
groupedByYear
```

```python
ax = groupedByYear.plot(marker = 's')
ax.set_title('Houston Rockets’ FieldGoals, 3PointShots and FreeThrows success rates')
ax.set_xlabel('Years')
ax.set_ylabel('Success Rate')
ax.set_xticks([2014,2015, 2016,2017,2018])
```

```python
#Q11
winData = nba[nba['WINorLOSS'] == 'W']
groupedWinByYear = nba.groupby(['Team', nba['Date'].dt.year])
groupedWinByYear.size()
```

```python
houWinData = winData[winData['Team'] == 'HOU']
houWinData.groupby(houWinData['Date'].dt.year).size().plot.bar(color='red')
```

```python
gswWinData = winData[winData['Team'] == 'GSW']
gswWinData.groupby(gswWinData['Date'].dt.year).size().plot.bar(color='yellow')
```

```python
#Q12
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
houWinData.groupby(houWinData['Date'].dt.year).size().plot.bar(color='red', ax=axes[0])
gswWinData.groupby(gswWinData['Date'].dt.year).size().plot.bar(color='yellow', ax=axes[1])
axes[0].set_title('Number of Wins of Houston Rockets')
axes[1].set_title('Number of Wins of Golden State Warriors')
axes[0].set_xlabel('Year')
axes[1].set_xlabel('Year')
```
