```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
#run the command to make matplotlib charts inline in the notebook
%matplotlib inline
```

```python
store=pd.read_excel('superstoreDataset.xlsx',index_col=0)
store.head()
```

```python
store.isna().any()
```

```python
store[store['Product Base Margin'].isnull()]
```

```python
totalsales=store['Sales'].groupby((store['State'],store['Product Category'],store['Product Sub-Category'])).sum()
```

```python
totalsales
```

```python
Texas=store[store['State']=='Texas']
txprofit=Texas['Profit'].groupby(Texas['Product Sub-Category']).sum()
```

```python
txprofit
```

```python
ax=txprofit.plot.bar()
ax.set_title('Texas Sales Profit Analysis')
ax.set_xlabel('Product Sub-Category')
ax.set_ylabel('Profit')
```

```python
sales=store.pivot_table('Sales',index=store['Order Date'].dt.year,columns='State',aggfunc='sum')
```

```python
sales
```

```python
ax=sales.plot(figsize=(10,5))
ax.set_title('Total sales of each state from 2009 to 2012')
ax.set_xlabel('Year')
ax.set_ylabel('States')
```

```python
discount1=store.pivot_table('Discount',index=store['Order Date'].dt.year,columns='Product Category',aggfunc='mean')
```

```python
discount2=store.pivot_table('Discount',index=store['Order Date'].dt.year,columns='Customer Segment',aggfunc='mean')
```

```python
fig, ax=plt.subplots(1,2,figsize=(10,3))
discount1.plot.bar(ax=ax[0])
discount2.plot.bar(ax=ax[1])
ax[0].set_title('Discount Rates by Product Category')
ax[1].set_title('Discount Rates by customer Segment')
ax[0].set_xlabel('year')
ax[1].set_xlabel('year')
ax[0].set_ylabel('Discount rate')
ax[1].set_ylabel('Discount rate')
```
