```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
#run the command to make matplotlib charts inline in the notebook
%matplotlib inline
```

```python
Crime=pd.read_csv('HoustonCrime.csv',index_col='OBJECTID')
Crime.head()
```

```python
Crime.dropna(subset=['X','Y','Time_Begun','HPD_Division','Premise_Type','Zip_Code'],inplace=True)
Crime.count()
```

```python
Crime['Zip_Code']=Crime['Zip_Code'].astype(int).astype(str)
```

```python
Crime.head()
```

```python
#we want to get first ten char from time_begun
Crime['Time_Begun']=Crime['Time_Begun'].str[:10]
#next, I create a list of unique da
dates=Crime['Time_Begun'].unique()
dates.sort()
dates
```

```python
#now we want to sort column in time_begun and convert it into datetime
Crime.sort_values(by='Time_Begun',inplace=True)
from datetime import datetime 
Crime['Time_Begun']=[datetime.strptime(x,'%Y-%m-%d')
                    for x in Crime['Time_Begun']]
```

```python
Crime['Time_Begun'].dtype
```

```python
#group Crime by date
Crime1=Crime[Crime['Offense']=='Theft']
ax=Crime1.groupby(Crime['Time_Begun'].dt.date).size().plot(figsize=(12,8),marker='s',label='A')
ax.set_xlabel('Date')
ax.set_ylabel('Num of cases')
ax.set_title('Number of cases by date',size=20)
ax.legend(loc='lower left')

```

```python
Crime.pivot_table(['Incident_No'],index='HPD_Division',columns='Offense',aggfunc='count',margins=True)
```

```python
offense_burg=Crime[Crime['Offense']=='Burglary']
ax=offense_burg.groupby(['Zip_Code']).size()
ax=ax.plot.bar(figsize=(12,8))
ax.set_xlabel('Zip_Code')
ax.set_ylabel('No. of Cases')
ax.set_title('Number of burglary cases by Zip_Code',size=20)
```
